```python
import numpy as np

# Fixing the seed for reproducibility
np.random.seed(0)

# Mean vector (-3, 0)
mean = [-3, 0]

# Covariance matrix
cov_matrix = [[1.0, 0.8], [0.8, 1.0]]

# Generating 500 random numbers with multivariate normal distribution
random_numbers = np.random.multivariate_normal(mean, cov_matrix, 500)

# Checking the shape
print(random_numbers.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(random_numbers[:, 0], random_numbers[:, 1], alpha=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

```python
# Criar os histogramas
plt.figure(figsize=(12, 5))

# Histogram for the first dimention(X)
plt.subplot(1, 2, 1)
plt.hist(random_numbers[:, 0], bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel("X")
plt.ylabel("Frequency")
plt.title("histogram of X")

# Histogram for the second dimention(Y)
plt.subplot(1, 2, 2)
plt.hist(random_numbers[:, 1], bins=30, alpha=0.7, color='green', edgecolor='black')
plt.xlabel("Y")
plt.ylabel("Frequency")
plt.title("Histogram of Y")

plt.xlim(-6, 3)

# show grafics
plt.tight_layout()
plt.show()
```

```python
mean2 = [0, -3]
random_numbers2 = np.random.multivariate_normal(mean2, cov_matrix, 500)

plt.figure(figsize=(8, 6))
plt.scatter(random_numbers[:, 0], random_numbers[:, 1], color='blue', alpha=0.5, label="Problem 1 (-3,0)")
plt.scatter(random_numbers2[:, 0], random_numbers2[:, 1], color='red', alpha=0.5, label="Problem 4 (0,-3)")

# Configurações do gráfico
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of the two data sets")
plt.legend()
plt.grid(True)

plt.show()
```

```python
combined_data = np.vstack((random_numbers, random_numbers2))  
print(combined_data.shape)
```

```python
labels1 = np.zeros((500, 1))  # 500 zeros
labels2 = np.ones((500, 1))   # 500 ones

random_numbers_labeled = np.hstack((random_numbers, labels1)) 
random_numbers2_labeled = np.hstack((random_numbers2, labels2))


combined_data = np.vstack((random_numbers_labeled, random_numbers2_labeled))

print(combined_data) 
```

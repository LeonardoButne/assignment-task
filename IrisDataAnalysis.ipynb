```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Carregar o dataset IRIS
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target

# Mapeia os números das espécies para os nomes das flores
species_names = {0: "Setosa", 1: "Versicolor", 2: "Virginica"}
df["species"] = df["species"].map(species_names)

#1️⃣ Create a scatter plot for two specific characteristics
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue="species", data=df)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()

#2️⃣ Create a scatterplot matrix (pairplot)
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.show()

#3️⃣ Create a correlation coefficient matrix
correlation_matrix = df.drop("species", axis=1).corr()
print("Correlation Matrix:")
print(correlation_matrix)

#4️⃣ Create a correlation matrix heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heat Map")
plt.show()
```

Analysis of the Obtained Results
After creating the charts and tables, we can draw several conclusions about the Iris dataset.

1️⃣ Scatter Plot: Relationship Between Sepal Length and Sepal Width
✅ Observation:

The species Setosa, Versicolor, and Virginica have distinct distributions.
Setosa stands out clearly since its sepal length and width values are well separated from the other two species.
Versicolor and Virginica have more overlap, making it harder to distinguish them using only these two features.
📝 Conclusion:

Sepal Width is not a strong criterion for separating Versicolor and Virginica, but it effectively distinguishes Setosa.
2️⃣ Pairplot: Scatter Matrix of All Features
✅ Observation:

Petal Length and Petal Width show a clear separation between the three species.
Setosa has significantly smaller petals, while Versicolor and Virginica have longer ones.
Versicolor has intermediate values, whereas Virginica has the largest petal measurements.
📝 Conclusion:

Petal Length and Petal Width are the most effective features for distinguishing the species.
Sepal Length and Sepal Width overlap for some species, making separation more difficult.
3️⃣ Correlation Matrix
✅ Observation:

Petal Length and Petal Width have a strong positive correlation (~0.96).
Sepal Width has a weak correlation with other variables.
Sepal Length and Petal Length also show a high correlation (~0.87), indicating that larger sepals tend to have larger petals.
📝 Conclusion:

Since Petal Length and Petal Width are highly correlated, one may be sufficient for classifying the species.
Sepal Width appears to be the least useful feature for species separation.
4️⃣ Heatmap of the Correlation Matrix
✅ Observation:

The colors indicate strong or weak relationships between features.
The strong red hue between Petal Length and Petal Width confirms their high positive correlation.
The weak relationship between Sepal Width and other variables reinforces the idea that it is not a good criterion for species differentiation.
📝 Conclusion:

Petal Length and Petal Width are the best attributes for differentiating species.
Sepal Width has little influence on species classification.
General Conclusion
🚀 The best features for classifying Iris species are:
✅ Petal Length
✅ Petal Width

🔴 The least useful feature is:
❌ Sepal Width

💡 If we were to build a classifier, we would likely use only petal measurements!

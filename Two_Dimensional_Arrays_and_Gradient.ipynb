```python
import numpy as np
import matplotlib.pyplot as plt

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    
    x = np.arange(*x_range)
    y = function(x)
    gradient = np.diff(y) / np.diff(x)  # Calculation of the derivative
    array_xy = np.column_stack((x, y))  # Combining x and y into a matrix
    return array_xy, gradient

def function1(x):
    return x**2

def function2(x):
    return 2*x**2 + 2*x

def function3(x):
    return np.sin(x / 12)

x_range1 = (-50, 50.1, 0.1)
x_range3 = (0, 50.1, 0.1)

array_xy1, gradient1 = compute_gradient(function1, x_range1)
array_xy2, gradient2 = compute_gradient(function2, x_range1)
array_xy3, gradient3 = compute_gradient(function3, x_range3)

# Creating the graphs
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

#Plot of function 1 and its gradient
axs[0, 0].plot(array_xy1[:, 0], array_xy1[:, 1], label="y = x²", color='blue')
axs[0, 0].set_title("Function 1")
axs[0, 0].legend()
axs[0, 0].grid()

axs[0, 1].plot(array_xy1[:-1, 0], gradient1, label="Gradient", color='red')
axs[0, 1].set_title("Gradient of Function 1")
axs[0, 1].legend()
axs[0, 1].grid()

# Plot of function 2 and its gradient
axs[1, 0].plot(array_xy2[:, 0], array_xy2[:, 1], label="y = 2x² + 2x", color='blue')
axs[1, 0].set_title("Function 2")
axs[1, 0].legend()
axs[1, 0].grid()

axs[1, 1].plot(array_xy2[:-1, 0], gradient2, label="Gradient", color='red')
axs[1, 1].set_title("Gradient of Function 2")
axs[1, 1].legend()
axs[1, 1].grid()

# Plot of function 3 and its gradient
axs[2, 0].plot(array_xy3[:, 0], array_xy3[:, 1], label="y = sin(x/12)", color='blue')
axs[2, 0].set_title("Function 3")
axs[2, 0].legend()
axs[2, 0].grid()

axs[2, 1].plot(array_xy3[:-1, 0], gradient3, label="Gradient", color='red')
axs[2, 1].set_title("Gradient of Function 3")
axs[2, 1].legend()
axs[2, 1].grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate gradient and find minimum
def compute_minimum(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)  # Criar array de x
    y = function(x)          # Calcular y
    
    min_value = np.min(y)          # Encontrar o menor valor de y
    min_index = np.argmin(y)       # Índice do menor valor
    min_x = x[min_index]           # Valor de x correspondente ao mínimo
    
    # Calculate gradient using finite differences
    gradient = np.diff(y) / np.diff(x)
    
    # Gradient before and after minimum
    gradient_before = gradient[min_index - 1] if min_index > 0 else None
    gradient_after = gradient[min_index] if min_index < len(gradient) else None

    return min_x, min_value, gradient_before, gradient_after

# Applying to functions
functions = [function1, function2, function3]
function_names = ["y = x^2", "y = 2x^2 + 2x", "y = sin(x/12)"]

for i, func in enumerate(functions):
    min_x, min_y, grad_before, grad_after = compute_minimum(func)
    print(f"Function: {function_names[i]}")
    print(f"Minimum y: {min_y} at x = {min_x}")
    print(f"Gradient before: {grad_before}, Gradient after: {grad_after}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt

days = 100
rice_per_day = [2**i for i in range(days)]  # List of rice grains per day
cumulative_rice = np.cumsum(rice_per_day)  # Cumulative sum of rice

rice_on_day_100 = rice_per_day[-1]


print(f"Rice on day 100: {rice_on_day_100}")
print(f"Total rice after 100 days: {cumulative_rice[-1]}")

# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, days + 1), rice_per_day, label='Rice per Day')
plt.plot(range(1, days + 1), cumulative_rice, label='Cumulative Rice', linestyle='--')
plt.xlabel('Day')
plt.ylabel('Number of Rice Grains')
plt.title('Rice Grains on Each Day and Cumulative Total')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    rice_per_day = [1]
    total_rice = [1]
    
    for day in range(1, n_days):
        rice_per_day.append(rice_per_day[day-1] * 2)
        total_rice.append(total_rice[day-1] + rice_per_day[day])
    
    print(f"Total grains on the chosen day: {rice_per_day [-1]}")
    print(f"Total grains up to the chosen day: {total_rice[-1]}")
    
    return rice_per_day, total_rice

# Usando a função para calcular os resultados para 10 dias
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=10)

# Plotando os resultados
plt.plot(range(1, 11), list_n_grains, label="Rice grains per day")
plt.plot(range(1, 11), list_total_grains, label="Total Cumulative")
plt.xlabel('Day')
plt.ylabel('Number of Rice Grains')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def calculate_days_of_rice(grains, people):
    # Assuming each person consumes 150 grains of rice per day
    grains_per_day_per_person = 150
    
    # Total rice needed per day for all people
    total_grains_needed_per_day = grains_per_day_per_person * people
    
    # Calculating how many days the rice will last
    days = grains // total_grains_needed_per_day
    return days
```
